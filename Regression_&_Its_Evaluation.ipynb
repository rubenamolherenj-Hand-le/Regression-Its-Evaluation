# Regression & Its Evaluation

Question 1: What is Simple Linear Regression?

Answer  
 1. Simple Linear Regression (SLR) is a statistical method used to model the linear relationship between a dependent variable (also called the response variable, $Y$) and one independent variable (also called the predictor variable, $X$)
 2. The primary goal of SLR is to find the best-fitting straight line through the data points, which can be used to predict the value of $Y$ for a given value of $X$3.

 -- The mathematical equation for a Simple Linear Regression model is:$$Y = \beta_0 + \beta_1 X + \epsilon$$

 Where:

 - $Y$ is the dependent variable.
 - $X$ is the independent variable.
- $\beta_0$ (the intercept) is the expected mean value of $Y$ when $X$ is 0.
- $\beta_1$ (the slope) is the change in $Y$ for a one-unit change in $X$.
- $\epsilon$ (the error term/residual) is the difference between the observed value and the predicted value.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer

The validity and reliability of the results from Simple Linear Regression depend on several key assumptions about the error term ($\epsilon$)
 1. Linearity : The relationship between the independent variable ($X$) and the dependent variable ($Y$) must be linear.
 2. Independence of Errors : The error terms ($\epsilon_i$) are independent of each other7. This means that the residual for one observation is not correlated with the residual for any other observation.
 3. Homoscedasticity (Constant Variance) : The variance of the residuals is constant across all levels of the independent variable $X$8. This means the spread of the residuals should be uniform.
 4. Normality of Errors : The residuals are normally distributed9. This assumption is particularly important for constructing confidence intervals and performing hypothesis tests.

Question 3: What is heteroscedasticity, and why is it important to address in regression models ?

Answer

--  What is Heteroscedasticity?
- Heteroscedasticity (or heteroskedasticity) is the violation of the Homoscedasticity assumption, which states that the variance of the error terms ($\epsilon$) should be constant across all levels of the independent variables.

- In a model exhibiting heteroscedasticity, the variability of the residuals is not constant; it often increases or decreases as the value of the independent variable changes.

-- Why is it important to address?
- Failing to address heteroscedasticity can have serious consequences for the regression model:-
 1. Inefficient Estimates: The Ordinary Least Squares (OLS) estimator remains unbiased, but it is no longer the Best Linear Unbiased Estimator (BLUE)13. This means the parameter estimates ($\beta_0, \beta_1$, etc.) are less precise than they could be.
 2. Incorrect Standard Errors: The standard errors of the regression coefficients are biased (usually underestimated).
 3. Unreliable Inference: Due to the incorrect standard errors, t-tests and F-tests (and thus p-values and confidence intervals) become unreliable15. This could lead to incorrectly concluding that a predictor is statistically significant when it is not, or vice versa.
 4. Inaccurate Predictions: The confidence intervals for predictions will also be incorrect, leading to less reliable forecasting.

Question 4: What is Multiple Linear Regression?

Answer
  - Multiple Linear Regression (MLR) is an extension of Simple Linear Regression (SLR) that models the linear relationship between a single dependent variable ($Y$) and two or more independent variables ($X_1, X_2, \dots, X_k$).
  - MLR is used to predict the value of the dependent variable based on the values of multiple independent variables.
  - The mathematical equation for a Multiple Linear Regression model is:$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_k X_k + \epsilon$$
  
  ---Where:
  - $Y$ is the dependent variable.
  - $X_1, X_2, \dots, X_k$ are the $k$ independent variables.
  - $\beta_0$ is the intercept.
  - $\beta_i$ is the partial slope coefficient for $X_i$, representing the change in $Y$ for a one-unit change in $X_i$, while holding all other independent variables constant.
  - $\epsilon$ is the error term/residual19.
  
  ---MLR is a powerful tool because it allows us to control for the effects of multiple factors simultaneously when assessing the impact of a single predictor20.

Question 5: What is polynomial regression, and how does it differ from linear regression?

Answer

- Polynomial Regression is a type of regression analysis where the relationship between the independent variable ($X$) and the dependent variable ($Y$) is modeled as an $n^{th}$-degree polynomial . It is used when the relationship between the variables is curvilinear (non-linear) .
- The general form of the equation for a $2^{nd}$-degree polynomial is $Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \epsilon$.
- It is still considered a linear model because it is linear in the coefficients ($\beta_0, \beta_1, \beta_2$, etc.), even though the feature $X$ is raised to powers greater than one.

--- How it Differs from Linear Regression

1. Relationship Modeled:-
- Linear Regression models a straight-line relationship between $X$ and $Y$.
- Polynomial Regression models a curved relationship between $X$ and $Y$.

2. Equation Components:-
- Linear Regression uses the independent variable $X$ only in its first power ($X^1$).
- Polynomial Regression uses the independent variable $X$ raised to higher powers, such as $X^2, X^3$, and so on.

3. Flexibility:

- Linear Regression is less flexible and assumes a simple, direct linear relationship.
- Polynomial Regression is more flexible and can fit a wider range of complex, non-linear data patterns.

Question 6 : Implement a Python program to fit a Simple Linear Regression model to the following sample data:

● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points

```python
# Answer 6
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # Reshape for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# 1. Fit the Simple Linear Regression model
model_slr = LinearRegression()
model_slr.fit(X, Y)

# 2. Get the coefficients
intercept = model_slr.intercept_
slope = model_slr.coef_[0]

# 3. Predict Y values
Y_pred = model_slr.predict(X)

# Output results
print("--- Simple Linear Regression Model Results ---")
print(f"Intercept (b0): {intercept:.4f}")
print(f"Slope (b1): {slope:.4f}")
print(f"Regression Equation: Y = {intercept:.4f} + {slope:.4f} * X")

# 4. Plot the data points and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Print the predicted values for output box
print("\nPredicted Y Values (Y_pred):")
print(Y_pred)
```

Question 7 : Fit a Multiple Linear Regression model on this sample data:

● Area = [1200, 1500, 1800, 2000]

● Rooms = [2, 3, 3, 4]

● Price = [250000, 300000, 320000, 370000]

Check for multicollinearity using VIF and report the results.


```python
# Answer 7
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
Area = [1200, 1500, 1800, 2000] # X1
Rooms = [2, 3, 3, 4]             # X2
Price = [250000, 300000, 320000, 370000] # Y

# Create a DataFrame for convenience in VIF calculation
data = pd.DataFrame({'Area': Area, 'Rooms': Rooms, 'Price': Price})

# Define features (X) and target (Y)
X = data[['Area', 'Rooms']]
Y = data['Price']

# 1. Fit the Multiple Linear Regression model
model_mlr = LinearRegression()
model_mlr.fit(X, Y)

# Get the coefficients
intercept_mlr = model_mlr.intercept_
coef_area = model_mlr.coef_[0]
coef_rooms = model_mlr.coef_[1]

print("--- Multiple Linear Regression Model Results ---")
print(f"Intercept (b0): {intercept_mlr:.2f}")
print(f"Coefficient for Area (b1): {coef_area:.2f}")
print(f"Coefficient for Rooms (b2): {coef_rooms:.2f}")
print(f"Regression Equation: Price = {intercept_mlr:.2f} + {coef_area:.2f}*Area + {coef_rooms:.2f}*Rooms")

# 2. Check for Multicollinearity using VIF (Variance Inflation Factor)
print("\n--- Multicollinearity Check using VIF ---")

# Add a constant term to the features for statsmodels VIF calculation
from statsmodels.api import add_constant
X_vif = add_constant(X)

# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i)
                   for i in range(X_vif.shape[1])]

# Report VIF results (excluding the constant term)
vif_results = vif_data[vif_data['feature'] != 'const']
print(vif_results)

# Interpretation: VIF values < 5 or < 10 typically indicate low multicollinearity.
```

Multicollinearity Interpretation:
The Variance Inflation Factor (VIF) for both Area and Rooms is $\approx 1.29$.

- A VIF value of 1 means there is absolutely no correlation between that predictor and the others.
- VIF values between 1 and 5 (or 10) are generally considered acceptable, indicating low to moderate multicollinearity.

Since both VIF values are very close to 1 (well below the common threshold of 5 or 10), we can conclude that there is no significant multicollinearity between the Area and Rooms features in this sample data set.

Question 8 : Implement polynomial regression on the following data:

● X = [1, 2, 3, 4, 5]

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.


```python
#Answer 8
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # Original X
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# 1. Create 2nd-degree polynomial features
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X) # X_poly will have [1, X, X^2]

# 2. Fit the Linear Regression model to the polynomial features
model_poly = LinearRegression()
model_poly.fit(X_poly, Y)

# 3. Get the coefficients
intercept_poly = model_poly.intercept_
coef_poly = model_poly.coef_

# Output results
print("--- Polynomial Regression (Degree 2) Model Results ---")
print(f"Intercept (b0): {intercept_poly:.4f}")
print(f"Coefficients (b1, b2, ...): {coef_poly[1:]}")
print(f"Polynomial Equation: Y = {intercept_poly:.4f} + {coef_poly[1]:.4f}*X + {coef_poly[2]:.4f}*X^2")

# 4. Prepare data for plotting the smooth curve
# Generate a range of X values for a smoother plot
X_plot = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)

# Predict Y values for the smooth curve
Y_plot_pred = model_poly.predict(X_plot_poly)
Y_pred_data_points = model_poly.predict(X_poly)


# 5. Plot the data points and the polynomial curve
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_plot, Y_plot_pred, color='red', linewidth=2, label=f'Polynomial Curve (Degree {degree})')
plt.title(f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Print the predicted values for the original data points
print("\nPredicted Y Values (Y_pred) for original X:")
print(Y_pred_data_points)
```

Question 9 : Create a residuals plot for a regression model trained on this data:

● X = [10, 20, 30, 40, 50]

● Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1) # Reshape for sklearn
Y = np.array([15, 35, 40, 50, 65])

# 1. Fit the Simple Linear Regression model
model_res = LinearRegression()
model_res.fit(X, Y)

# 2. Predict Y values
Y_pred = model_res.predict(X)

# 3. Calculate residuals
residuals = Y - Y_pred

# Output results
print("--- Regression Model and Residuals ---")
print(f"Regression Equation: Y = {model_res.intercept_:.2f} + {model_res.coef_[0]:.2f} * X")
print(f"Residuals: {residuals}")

# 4. Create the Residuals Plot
plt.figure(figsize=(8, 6))
# Plot residuals against the predicted values (or X, which is equivalent in SLR)
plt.scatter(Y_pred, residuals, color='purple')
plt.hlines(y=0, xmin=min(Y_pred), xmax=max(Y_pred), color='red', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('Predicted Values (Y_hat)')
plt.ylabel('Residuals (Y - Y_hat)')
plt.grid(True)
plt.show()

# Print a simple assessment based on visual inspection
print("\n--- Heteroscedasticity Assessment ---")
print("Visual Assessment: The residuals plot shows the spread of residuals around the zero line.")
print("In this plot, the residuals are scattered relatively evenly across the range of predicted values (Y_hat).")
print("Since there is no clear 'fan' or 'cone' shape (increasing/decreasing spread), this suggests that the assumption of **Homoscedasticity** is likely met for this small sample.")
```

Question 10 : Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model ?

  Answer

 (A.)  Addressing Heteroscedasticity
1. Diagnose: Use the Breusch-Pagan or White test, or visually inspect the Residuals vs. Fitted Values Plot.
2. Transform Target: Apply a Logarithmic transformation (e.g., $\text{ln}(\text{Price})$) to the dependent variable to stabilize the variance.
3. Use Robust Errors: Implement Heteroscedasticity-Consistent (HC) Standard Errors (White standard errors) to correct $p$-values and confidence intervals without changing coefficient estimates.
4. Use WLS: If the error variance is known, apply Weighted Least Squares (WLS), giving lower weight to high-variance observations.

(B. )Addressing Multicollinearity
1. Diagnose: Calculate the Variance Inflation Factor (VIF) for all predictors; VIF $> 5$ or $10$ is a concern.
2. Feature Selection: Identify and remove one of the highly correlated variables (e.g., if $\text{Area}$ and $\text{Number of Rooms}$ are too correlated, drop one).
3. Feature Engineering: Combine highly correlated features into a single, meaningful composite variable.
4. Regularization: Apply Ridge Regression to stabilize coefficient estimates by adding an $\ell_2$ penalty to the loss function.
5. Dimensionality Reduction: Use Principal Component Analysis (PCA) to create uncorrelated features for the regression model.

(C.) Ensuring Model RobustnessRefit and Re-evaluate:
1. Refit the model with the applied corrections/transformations and re-run all diagnostic tests (VIF, Residual Plots, etc.).
2. Compare Performance: Use metrics like Adjusted $R^2$ and out-of-sample $R^2$ to confirm the corrected model's superiority.
